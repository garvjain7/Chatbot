```python
import cv2
from google.colab.patches import cv2_imshow #for image display
```

```python
img=cv2.imread('drawing.jpg') #imread is for reading the images using opencv
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.subplot(141) #141 => 1->row(screen width) 4->column(screen height) 1->position of image
plt.imshow(img) #imshow is to show images

plt.subplot(142)
plt.imshow(img)

plt.subplot(143)
plt.imshow(img)

plt.subplot(144)
plt.imshow(img)
```

```python
print(img.size)
print(img.shape)
```

```python
cv2_imshow(img)
```

```python
r_image = img[:,:,0]
g_image = img[:,:,1]
b_image = img[:,:,2]
plt.subplot(221)
plt.imshow(r_image)
plt.subplot(222)
plt.imshow(g_image)
plt.subplot(223)
plt.imshow(b_image)
```

```python
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # to change image into rgb
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # to change image to gray scale
plt.subplot(221)
plt.imshow(rgb)
plt.subplot(222)
plt.imshow(gray)
```
