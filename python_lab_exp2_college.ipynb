```python
my_list = [1, 5, 2, 8, 5, 3, 5]
print(f"Original list: {my_list}")

print()
# count() method: Returns the number of times a specified value appears in the list.
count_of_5 = my_list.count(5)
print(f"Number of 5s in the list: {count_of_5}")

print()
# sort() method: Sorts the list in ascending order by default.
my_list.sort()
print(f"Sorted list (ascending): {my_list}")

print()
# Sort in descending order
my_list_desc = [1, 5, 2, 8, 5, 3, 5] # Re-initialize for demonstration
my_list_desc.sort(reverse=True)
print(my_list_desc)

print()
# reverse() method: Reverses the order of the elements in the list.
my_list_rev = [1, 5, 2, 8, 5, 3, 5] # Re-initialized for the desired result
my_list_rev.reverse()
print(my_list_rev)

print()
# copy() method: Returns a shallow copy of the list.
original_list = [1, 2, 3]
copied_list = original_list.copy()
print(f"Original list: {original_list}, Copied list: {copied_list}")

# Demonstrate that they are separate objects
copied_list.append(4)
print(f"Original list after appending to copy: {original_list}")
print(f"Copied list after appending to copy: {copied_list}")
print()
```

```python
# soft copy of list
import copy

# 1. Original list with a nested list
original = [1, 2, [3, 4]]

# --- CREATING COPIES ---

# Methods for Shallow Copy (Soft Copy)
shallow1 = original.copy()        # Built-in method
shallow2 = list(original)         # Constructor
shallow3 = original[:]            # Slicing

# Method for Deep Copy (Hard Copy)
deep_copy = copy.deepcopy(original)

# --- DEMONSTRATING APPENDS ---

# A. Appending to the TOP LEVEL
# This affects ONLY the list you append to.
original.append(5)

# B. Appending to the NESTED LIST (The "Soft" part)
# Because shallow copies share the same inner list reference,
# changing the inner list in ONE affects the original and ALL shallow copies.
shallow1[2].append("SHALLOW_CHANGE")

# C. Appending to the DEEP COPY
# This is a "Hard" copy; it is 100% independent.
deep_copy[2].append("DEEP_CHANGE")

# --- RESULTS ---
print(f"Original: {original}")
print(f"Shallow 1: {shallow1}  <-- Reflects nested change")
print(f"Shallow 2: {shallow2}  <-- Reflects nested change")
print(f"Deep Copy: {deep_copy} <-- Completely independent")
```

```python
# Write a program to find the no. is prime or not from a list
def is_prime(n):
  if n < 2: return False
  for i in range(2, int(n*0.5)+1):
    if n % i == 0: return False
  return True

numbers = [2, 7, 10, 15, 19, 21, 23]
for num in numbers:
  if is_prime(num):
    print(f"{num} is a prime number")
  else:
    print(f"{num} is not a prime number")
```

```python
# Write a program to find the no. is palindrome or not from a list
def isPalindrome(n):
  return str(n) == str(n)[::-1]

number = [121, 505, 123, 444, 987, 1001]
for num in number:
  if isPalindrome(num):
    print(f"{num} is a palindrome")
  else:
    print(f"{num} is not a palindrome")
```

```python
# Write a program to find the no. in list is armstrong or not
numbers = [153, 370, 371, 407, 123, 9474]

def isArmstrong(n):
    s = str(n)
    power = len(s)
    total = sum(int(digit)**power for digit in s)
    return total == n

print("Armstrong Check:")
for num in numbers:
    if isArmstrong(num):
        print(f"{num} is an Armstrong number")
    else:
        print(f"{num} is Not an Armstrong number")
```

### Set Operations Demonstrations

```python
# Initializing sets for demonstration
set_a = {1, 2, 3, 4, 5,4}
set_b = {4, 5, 6, 7, 8}
set_c = {1, 2, 3, 1, 2, 3}

print(set_a)
print(set_b)
print(set_c)
print()

# 1. add() method: Adds an element to the set.
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)
```

```python
# 2. copy() method: Returns a shallow copy of the set.
original_set = {10, 20, 30}
copied_set = original_set.copy()
print(original_set)
print(copied_set)
# Demonstrate independence
copied_set.add(40)
print(original_set)
print(copied_set)
```

```python
# 3. difference() method: Returns a set containing the difference between two or more sets.
diff_set = set_a.difference(set_b)
print(diff_set)
```

```python
# 4. discard() method: Removes the specified item from the set.
set_a_copy_for_discard = set_a.copy() # Use a copy to not alter original set_a for subsequent examples
set_a_copy_for_discard.discard(3)
print(f"Set A after discard(3): {set_a_copy_for_discard}")
set_a_copy_for_discard.discard(99) # Discarding a non-existent item does not raise an error
print(f"Set A after discard(99) (non-existent): {set_a_copy_for_discard}")

print("\n---")
```

```python
# 5. intersection() method: Returns a set, that is the intersection of two or more sets.
intersection_set = set_a.intersection(set_b)
print(f"Intersection of A and B: {intersection_set}")

print("\n---")
```

```python
# 6. issubset() method: Returns True if all items in the set exists in the specified set, otherwise False.
print(f"Is Set C a subset of Set A? {set_c.issubset(set_a)}")
print(f"Is Set A a subset of Set C? {set_a.issubset(set_c)}")

print("\n---")
```

```python
# 7. pop() method: Removes an arbitrary element from the set and returns it.
# Note: Sets are unordered, so you don't know which item will be removed.
pop_set = {100, 200, 300, 400}
popped_item = pop_set.pop()
print(f"Set before pop(): {{100, 200, 300, 400}}")
print(f"Popped item: {popped_item}")
print(f"Set after pop(): {pop_set}")

print("\n---")
```

```python
# 8. remove() method: Removes the specified item from the set.
# Raises an error if the item does not exist.
set_a_copy_for_remove = set_a.copy()
set_a_copy_for_remove.remove(2)
print(f"Set A after remove(2): {set_a_copy_for_remove}")

# Uncomment the following lines to see the error for removing a non-existent item:
# try:
#     set_a_copy_for_remove.remove(99)
# except KeyError as e:
#     print(f"Error when removing non-existent item: {e}")

print("\n---")
```

```python
# 9. union() method: Returns a set containing all items from both sets.
union_set = set_a.union(set_b)
print(f"Union of A and B: {union_set}")
```

```python
# 10. update() method: Inserts the items from another set into the current set.
set_a_for_update = {1, 2, 3}
set_a_for_update.update(set_b)
print(f"Set A after update with Set B: {set_a_for_update}")

print("\n---")
```

## Summary of Dictionary Operations

This section summarizes the dictionary operations and iteration methods demonstrated above, highlighting their uses and characteristics.

### Dictionary Methods:

*   **`clear()`**: Removes all key-value pairs from the dictionary, making it empty. Useful for resetting a dictionary.

*   **`copy()`**: Returns a shallow copy of the dictionary. Changes to the copied dictionary's top-level keys do not affect the original, but changes to nested mutable objects *will* affect both.

*   **`get(key, default)`**: Safely retrieves the value associated with a `key`. If the `key` is not found, it returns `None` by default, or a specified `default` value, preventing `KeyError`.

*   **`items()`**: Returns a view object that displays a list of a dictionary's key-value tuple pairs. Ideal for iterating over both keys and values simultaneously.

*   **`keys()`**: Returns a view object that displays a list of all the keys in the dictionary. Useful for iterating or checking for key existence.

*   **`pop(key, default)`**: Removes the item with the specified `key` and returns its value. If the `key` is not found and no `default` is provided, it raises a `KeyError`. If `default` is provided, it returns `default`.

*   **`popitem()`**: Removes and returns the last inserted key-value pair as a tuple. (In Python versions < 3.7, it removes and returns an arbitrary item).

*   **`setdefault(key, default)`**: Gets the value of a specified `key`. If the `key` does not exist, it inserts the `key` with the specified `default` value (or `None` if `default` is not provided) and returns the value. It *does not* overwrite an existing value.

*   **`update(iterable)`**: Updates the dictionary with elements from another dictionary object or from an iterable of key-value pairs. Existing keys are updated, and new keys are added.

*   **`values()`**: Returns a view object that displays a list of all the values in the dictionary. Useful for iterating or checking for value existence.

### Dictionary Iteration:

*   **Simple Loop (over keys)**: Iterating directly over a dictionary (e.g., `for key in my_dict:`) loops through its keys.

*   **Medium Loop (over key-value pairs)**: Using `my_dict.items()` allows for convenient iteration over both keys and values (e.g., `for key, value in my_dict.items():`).

*   **Hard Loop (Nested Structures & Conditional Logic)**: Involves iterating through dictionaries that contain nested data structures (like other dictionaries or lists) and applying conditional logic (e.g., `if` statements) to filter or process data based on specific criteria within these nested structures.
