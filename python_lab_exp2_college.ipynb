```python
# identity operators in python
x1=['apple','banana']
y1=['apple','banana']
z=x1
print(x1 is z) # True because x and z are the same object
print(x1 is y1) # False because x and ya re not the same objects even if they have same values
print(x1==y1) # True because '==' checks the elements are equal or not on each index
print(z is x1) # same reason as x and z are same objects
```

```python
# membership operator in python
print('pineapple' not in x1) # pineapple is not in list x.
print('banana' in y1) # banana is in list y.
```

```python
# IF-ELSE Statement in python
a,b = 33,33
if b > a:
  print("b is greater than a")
elif b < a:
  print("b is smaller than a")
else:
  print("a and b are equal")
```

```python
# Nested IF-ELSE Statements in python
x=41
if x>10:
  print("Above 10")
  if x>20:
    print("Above 20")
  else:
    print("but not above 20")
```

```python
# WHILE LOOP in pyton
i=1
while i<6:
  print(i)
  i=i+1 # i++ is not valid in python

print()
# For loop in python (loop through a list)
fruits = ['apple', 'banana', 'cherry']
for x in fruits:
  print(x)

print()
# Loop through the letters in word 'banana'
for x in 'banana':
  print(x)
```

```python
# break statement
fruits = ['apple', 'banana', 'cherry']
for x in fruits:
  print(x)
  if x == 'banana':
    break

print()
# range
for j in range(1,5):
  print(j)
```
